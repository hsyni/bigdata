## Read the Data into Pandas DataFrame without Unnamed Index Column

```python
import pandas as pd
df = pd.read_csv("poems_Kazan.csv", index_col = [0])
df
```

## Remove All Punctuation Characters in the DataFrame 

```python
import string

def remove_punc_generator(string):
    punc = '''!()-[]{};:'"\,<>./?@#$%^'&"*_’~“'''
    for ele in string:  
        if ele in punc:  
            string = string.replace(ele, "") 
    return string
```

```python
df.title = df.title.apply(remove_punc_generator)
df.poet = df.poet.apply(remove_punc_generator)
df.poem = df.poem.apply(remove_punc_generator)
df.poem[86411]
```

## Remove All Numbers Characters in the Poem 

```python
import re

def remove_number_generator(string):
    return re.sub(r'\d+', ' ', string)
```

```python
df['poem'] = df['poem'].apply(remove_number_generator)
df.poem[86411]
```

## Remove Multiple Space Characters in DataFrame

```python
df.title = df.title.replace(r'\s+', ' ', regex=True)
df.poet = df.poet.replace(r'\s+', ' ', regex=True)
df.poem = df.poem.replace(r'\s+', ' ', regex=True)
df.poem[86411]
```

## Remove One Letter Words

```python
df.poem = df.poem.map(lambda x: ' '.join(word for word in x.split() if len(word)>1))
df.poem[86411]
```

```python
!pip install zemberek-python

import time
import logging

from zemberek import (
    TurkishSpellChecker,
    TurkishSentenceNormalizer,
    TurkishSentenceExtractor,
    TurkishMorphology,
    TurkishTokenizer
)
```

## Create logger and morphology Objects

```python
logger = logging.getLogger(__name__)
morphology = TurkishMorphology.create_with_defaults()
```

##Remove Months

```python
def remove_months(row):
    poem = row['poem']
    poet = row['poet']
    
    # Check if poet value exists in poem column
    if poet in poem:
        # Delete poet value from poem column
        poem = poem.replace(poet, '')
    
    # Remove Turkish month names
    turkish_months = ['ocak', 'şubat', 'mart', 'nisan', 'mayıs', 'haziran',
                      'temmuz', 'ağustos', 'eylül', 'ekim', 'kasım', 'aralık', 'Ocak', 'Şubat', 'Mart', 'Nisan', 'Mayıs', 'Haziran',
                      'Temmuz', 'Ağustos', 'Eylül', 'Ekim', 'Kasım', 'Aralık']
    for month in turkish_months:
        poem = poem.replace(month, ' ')
    
    row['poem'] = poem
    return row

# Apply the remove_numbers_and_months function to the DataFrame
df = df.apply(remove_months, axis=1)
```

##Remove Days Names

```python
def remove_days(row):
    poem = row['poem']
    poet = row['poet']
    
    # Remove Turkish month names
    turkish_days = ["Pazartesi", "Salı", "Çarşamba", "Perşembe", "Cuma", "Cumartesi", "Pazar","pazartesi", "salı", "çarşamba", "perşembe", "cuma", "cumartesi", "pazar"]
    for day in turkish_days:
        poem = poem.replace(day, ' ')
    
    row['poem'] = poem
    return row

# Apply the remove_numbers_and_months function to the DataFrame
df = df.apply(remove_days, axis=1)
```

## Tokenizer

```python
tokenizer = TurkishTokenizer.DEFAULT
tokens = tokenizer.tokenize(df.poem[86411])
for token in tokens:
    print('Content = ', token.content)
    print('Type = ', token.type_.name)
    print('Start = ', token.start)
    print('Stop = ', token.end, '\n')
```

## Single Word Morphological Analysis

```python
for token in tokens:
    results = morphology.analyze(token.content)
    for result in results:
        print(result)
    print("\n")
```

## Sentence Extractor

```python
extractor = TurkishSentenceExtractor()
start = time.time()
logger.info(f"Sentence extraction instance created in: {time.time() - start} s")
df['poem_sentences'] = df.poem.map(lambda x: extractor.from_paragraph(x))
logger.info(f"Sentences extraction ended in: {time.time() - start} s")
df.poem_sentences
```

## Sentence Analysis and Disambiguation

```python
i = 1
for sentence in df.poem_sentences[86411]:
    analysis = morphology.analyze_sentence(sentence)
    after = morphology.disambiguate(sentence, analysis)
    print("\n Sentence ", i)
    print("--------------")
    print("\n  Before disambiguation")
    for e in analysis:
        print(f"    Word = {e.inp}")
        for s in e:
            print("      ", s.format_string())
    print("\nAfter disambiguation")
    for s in after.best_analysis():
        print("   ", s.format_string())
    i = i+1
```

## Sentence Normalization 

```python
normalizer = TurkishSentenceNormalizer(morphology)
logger.info(f"Normalization instance created in: {time.time() - start} s")

start = time.time()
df['normalized_poem_sent'] = df.poem.map(lambda x: normalizer.normalize(x))
logger.info(f"Sentences normalized in: {time.time() - start} s")
df_normalized = df.normalized_poem_sent
df_normalized.to_csv('normalized_data3.csv')
```
